## Save hot & slow stars in merge_good, and cool or fast stars in merge_bad.

```python
m = pd.merge(smsyn, smemp, on=['obs'], suffixes=['_syn', '_emp'])#, how='outer')

merge_good = m.query('teff_emp > 4700 and vsini < 20')
merge_bad = m.query('not(teff_emp > 4700 and vsini < 20)')
specmatch_synth_names = merge_good.name
specmatch_emp_names = merge_bad.name

merge_good = merge_good.dropna(axis='columns', how='all')

#g = smsyn.groupby('name')
#df = g.first()[['obs']]

#teff = g.median()['teff'].sort_values()
#teff = teff.drop('0429-01631-1')
#df['teff'] = teff
#df
print(len(merge_good))
```
